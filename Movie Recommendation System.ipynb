# Movie Recommendation System

# Objective

**Recommender System** is a system that seeks to predict or filter preferences according to the user's choices. Recommender systems are utilized in a variety of areas including movies, music, news, books, research articles, search queries, social tags, and products in general. Recommender systems produce a list of recommendations in any of the two ways -

**Collaborative filtering:** Collaborative filtering approaches build a model from the user's past behavior (i.e. items purchased or searched by the user) as well as similar decisions made by other users. This model is then used to predict items (or ratings for items) that users may have an interest in.

**Content-based filtering:** Content-based filtering approaches uses a series of discrete characteristics of an item in order to recommend additional items with similar properties. Content-based filtering methods are totally based on a description of the item and a profile of the user's preferences. It recommends items based on the user's past preferences. Let's develop a basic recommendation system using Python and Pandas.

Let's develop a basic recommendation system by suggesting items that are most similar to a particular item, in this case, movies. It just tells what movies/items are most similar to the user's movie choice.

# Get Similarity Score using Cosine Similarity

cosine similarity computes the L2-normalized dot product of vectors. Euclidean (L2) normalization projects the vectors onto the unit sphere, and their dot product is then the cosine of the angle between the points denoted by the vectors.

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
Similarity_Score = cosine_similarity (X)
```

# Project Explanation: Movie Recommendation System

The **Movie Recommendation** System aims to provide users with personalized movie suggestions by predicting their preferences based on past choices. Recommender systems are widely used in various fields such as movies, music, e-commerce, news, and social media to enhance user experience by providing tailored recommendations. In this project, we will implement two widely-used recommendation techniques: **Collaborative Filtering** and **Content-Based Filtering.**

**Collaborative Filtering**
Collaborative filtering is a popular approach in recommender systems that leverages the collective behavior of users to make recommendations. It assumes that if users have agreed on certain items in the past, they are likely to agree on similar items in the future. Collaborative filtering operates in two main forms:

1. **User-User Collaborative Filtering:** This approach identifies users with similar tastes based on their interactions (e.g., ratings or purchases). For example, if User A and User B both liked the same set of movies, the system will recommend movies that User B liked to User A and vice versa.

2. **Item-Item Collaborative Filtering:** Instead of finding similar users, this approach focuses on finding items that are similar. If a user liked certain movies, the system will recommend other movies that are similar to those liked movies, based on the behavior of other users.

In this project, we build a collaborative filtering model using user interaction data (i.e., movie ratings). The model predicts the ratings of movies that a user hasn’t watched yet, based on the ratings from users with similar preferences.

**Content-Based Filtering**
Content-based filtering approaches recommendations by analyzing the attributes of the items and the preferences of users. This method focuses entirely on matching the characteristics of items with the user’s past interactions.

For instance, if a user has watched and enjoyed action movies, the system will recommend other movies that fall within the "action" genre. The key component here is a detailed profile of the user’s past preferences, which is used to recommend new items with similar properties. Features such as genre, cast, directors, and even user reviews can be taken into account in a content-based model.

In this system, we will extract features of movies such as genres and cast, and create a user profile based on the user’s past movie preferences. This profile will then be used to suggest new movies that match the user’s interests.
